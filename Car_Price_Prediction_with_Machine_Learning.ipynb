Checking Multicollinearity Using VIF

```python
import pandas as pd
import numpy as np
from statsmodels.stats.outliers_influence import variance_inflation_factor

# Assuming 'clean_data' is your DataFrame
# Fill missing values with mean or median, or use another appropriate method
clean_data.fillna(clean_data.mean(), inplace=True)

# Convert non-numeric values to numeric, if any
clean_data['Fuel_Type_Diesel'] = clean_data['Fuel_Type_Diesel'].astype('category').cat.codes
clean_data['Fuel_Type_Petrol'] = clean_data['Fuel_Type_Petrol'].astype('category').cat.codes
clean_data['Selling_type_Individual'] = clean_data['Selling_type_Individual'].astype('category').cat.codes
clean_data['Transmission_Manual'] = clean_data['Transmission_Manual'].astype('category').cat.codes

# Replace infinity values with NaNs and then fill them with appropriate values
clean_data.replace([np.inf, -np.inf], np.nan, inplace=True)
clean_data.fillna(clean_data.mean(), inplace=True)

# Selecting relevant variables for VIF calculation
variables = clean_data[['Present_Price','Driven_kms','Owner','Car_age','Fuel_Type_Diesel', 'Fuel_Type_Petrol',
                        'Selling_type_Individual','Transmission_Manual']]

# Calculating VIF
vif = pd.DataFrame()
vif["VIF"] = [variance_inflation_factor(variables.values, i) for i in range(variables.shape[1])]
vif["Features"] = variables.columns
print(vif)
```

Model Development

```python
X = final_df.drop('Selling_Price', axis=1)
y = final_df['Selling_Price']
```

Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(X[['Present_Price','Car_age']])
```

```python
input_scaled = scaler.transform(X[['Present_Price','Car_age']])
scaled_data = pd.DataFrame(input_scaled, columns=['Present_Price','Car_age'])
```

```python
X_scaled =scaled_data.join(X.drop(['Present_Price','Car_age'],axis=1))
```

Training and Testing Data

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(X_scaled,y,test_size=0.2, random_state=365)
```

Linear Regression Model

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()

# Training Model
lr.fit(x_train,y_train)

# Model Summary
y_pred_lr = lr.predict(x_test)

r_squared = r2_score(y_test,y_pred_lr)
rmse = np.sqrt(mean_squared_error(y_test,y_pred_lr))
print("R_squared :",r_squared)
```

Random Forest Regressor Model

```python
from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor()

# Training Model
rf.fit(x_train,y_train)

# Model Summary
y_pred_rf = rf.predict(x_test)

r_squared = r2_score(y_test,y_pred_rf)
rmse = np.sqrt(mean_squared_error(y_test,y_pred_rf))
print("R_squared :",r_squared)
```

GBT Regressor Model

```python
from sklearn.ensemble import GradientBoostingRegressor
gbt = GradientBoostingRegressor()

# Training Model
gbt.fit(x_train,y_train)

# Model Summary
y_pred_gbt = gbt.predict(x_test)

r_squared = r2_score(y_test,y_pred_gbt)
rmse = np.sqrt(mean_squared_error(y_test,y_pred_gbt))
print("R_squared :",r_squared)
```

Hyperparameter optimization using RandomizedSearchCV

```python
from sklearn.model_selection import RandomizedSearchCV
# Number of trees in random forest
n_estimators = [int(x) for x in np.linspace(start = 100, stop = 1200, num = 12)]
# Number of features to consider at every split
max_features = ['auto', 'sqrt']
# Maximum number of levels in tree
max_depth = [int(x) for x in np.linspace(5, 30, num = 6)]
# max_depth.append(None)
# Minimum number of samples required to split a node
min_samples_split = [2, 5, 10, 15, 100]
# Minimum number of samples required at each leaf node
min_samples_leaf = [1, 2, 5, 10]

# Create the random grid
random_grid = {'n_estimators': n_estimators,
               'max_features': max_features,
               'max_depth': max_depth,
               'min_samples_split': min_samples_split,
               'min_samples_leaf': min_samples_leaf}

print(random_grid)
```

```python
rf_random = RandomizedSearchCV(estimator = rf, param_distributions = random_grid, n_iter = 10, cv = 5, verbose=2, random_state=42, n_jobs = 1)
```

```python
rf_random.fit(x_train,y_train)
```

```python
rf_random.best_params_
```

```python
predictions=rf_random.predict(x_test)
r_squared = r2_score(y_test,predictions)
rmse = np.sqrt(mean_squared_error(y_test,predictions))
print("R_squared :",r_squared)
```

Test the model by giving input

```python
def prediction(PresentPrice, CarAge, FuelType,SellerType, Transmissionmanual):
    # feature scaling on present price and car age
    scaled_value = scaler.transform([[float(PresentPrice), int(CarAge)]])

    price = scaled_value[0,0]
    age = scaled_value[0,1]

    fuel = int(FuelType)
    Seller = int(SellerType)
    trans = int(Transmissionmanual)

    return rf_random.predict([[price, age, fuel, Seller, trans]])
```
